# InternetService

```python
 a = churn["InternetService"].value_counts()
```

```python
labels = 'Fiber optic', 'DSL', 'No'
explode = (0, 0, 0.1)
colors = sns.color_palette('bright')[1:4]
plt.figure(figsize=(10,10))
plt.pie(a, autopct="%.1f%%",explode = explode,labels = labels, colors = colors) #pctdistance=0.5)
plt.title("Internet service preffered by the customers ", fontsize=14)
plt.show()          
```

# Observations
* 44% of the customers preferring fiber optic internet
* 34.4% of the customers preferring dsl internet and 21.7% customers are not preferring internet.
* overall 78.4% of the customers are preferring internet.

# Multivariate analysis

### gender vs Totacharges vs churn

```python
ax = sns.barplot(x = churn.gender, y = churn.TotalCharges, hue = churn.Churn)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

# Observations
* There are total 2588 female customers paying total charges and they are not churn , whereas 1440 female customers are paying Totalcharges and they are  churn.
* 2511 male customers who are paying Totalcharges are not churn and 1623 customers who are paying Totalcharges and they are churn.

### gender vs SeniorCitizen vs churn

```python
ax = sns.barplot(x = churn.gender, y = churn.SeniorCitizen, hue = churn.Churn)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

## Observations
* More senior citizen female customers are churn compare. Same with male customers, more senior citizen male customers are churn.

### gender vs tenure vs churn

```python
sns.boxplot(x = churn.gender, y = churn.tenure, hue = churn.Churn)
```

## Observations
* Female customers who are not churn choosing tenure period of 15 - 60 days period(long tenure). Female customers who are churn are choosing the tenure period of 1 - 28 days(short tenure).
* Male customers who are not churn choosing tenure period of 15 - 60 days period(long tenure). Female customers who are churn are choosing the tenure period of 1 - 30 days(short tenure).

### contract vs Totalcharges vs churn

```python
ax = sns.barplot(x = churn.Contract, y = churn.TotalCharges, hue = churn.Churn)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
#sns.barplot(x = churn.Contract, y = churn.TotalCharges, hue = churn.Churn)
```

### Observations
* Customers whor are churn and choosing month to month contract are spending total charges of 1164 and not churn customers spending 1521.
* Customers who are churn and  choosing one year contract are spending total charges of 4066 and not churn customers spending 2901.
* Customers who are churn and choosing two year contract are spending total charges of 5432 and not churn customers spending 3655 respectively.
* customers who are choosig contract one year and two year and they are churn, spending more totalcharges. 

### InternetService vs tenure vs churn

```python
ax = sns.barplot(x = churn.InternetService, y = churn.tenure, hue = churn.Churn)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

### Observations
* Customers who are churn and choosing DSL internet service are preffering the tenure of 14 days and not churn customers preffering 37 days.
* Customers who are churn and choosing Fiberoptics internet service are preffering the tenure of 20 days and not churn customers preffering 42 days.
* Customers who are churn and not choosing internet service are preffering the tenure of 8 days and not churn customers preffering 32 days.
* Customers who are not churn customers are prefferring the more tenure and vice versa.
